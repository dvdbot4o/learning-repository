```python
"""Exercício 01
Faça um programa que peça dois números e imprima o maior deles."""

n1 = float(input(" digite um numero 1 "))

n2 = float(input(" digite um numero 2 "))

if n1 > n2:
  print(f" o numero {n1} é maior que o {n2}")
elif n1 < n2:
  print(f" o numero {n2} é maior que o {n1}")
else:
  print("os numeros sao iguais ")
```

```python
"""Exercício 06
Faça um programa que leia três números e mostre o maior deles:"""

numeros = []

for i in range(3):

    numeros.append(float(input(f" digite o numero {i + 1} ° ")))

print(numeros)

menor = numeros[0]

for i in numeros:
    if i < menor:
        menor = i
print(f" o menor numero da lista  é {menor}")
```

```python
"""Exercício 07
Faça um programa que leia três números e mostre o maior e o menor deles:"""

numeros = []

for i in range(3):

    numeros.append(float(input(f" digite o numero {i + 1} ° ")))

print(numeros)

menor = numeros[0]
maior = 0
for i in numeros:
    if i < menor:
        menor = i
    if i > maior:
        maior = i
print(f" o menor numero da lista  é {menor} e o maior é {maior}")

```

```python
"""

Exercício 08
Faça um programa que pergunte o preço de três produtos e informe qual produto você deve comprar, 
sabendo que a decisão é sempre pelo mais barato:

"""
numeros = []

for i in range(3):

    numeros.append(float(input(f" digite o valor do primeiro produto  {i + 1} ° ")))

print(numeros)

menor = numeros[0]
maior = 0
for i in numeros:
    if i < menor:
        menor = i
    if i > maior:
        maior = i
print(f" ess é o prduto mais barato custa R${menor:.2f}")
```
