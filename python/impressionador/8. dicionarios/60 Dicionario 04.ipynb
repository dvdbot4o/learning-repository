# For nos Dicionarios

### Estrutura:

```python
vendas_tecnologia = {
'iphone': 15000,
'samsung galaxy': 12000,
'tv samsung': 10000,
'ps5': 14300,
'tablet': 1720, 
'ipad': 1000,
'tv philco': 2500,
'notebook hp': 1000,
'notebook dell': 17000,
'notebook asus': 2450}

#demonstrando o for

for venda in vendas_tecnologia:
    print(venda)

for venda in vendas_tecnologia:
    print(f" item : {venda} unidades : {vendas_tecnologia[venda]}")
```

- Qual o total de notebooks vendidos?

```python
soma = 0

for venda in vendas_tecnologia:
    if "notebook" in venda:
        soma += vendas_tecnologia[venda]
print(soma)
```
